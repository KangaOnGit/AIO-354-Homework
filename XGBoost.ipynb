# **XGBoost For Regression**

---



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import xgboost as xgb
from sklearn.metrics import mean_squared_error, mean_absolute_error, accuracy_score
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
```

```python
lrl_df = pd.read_csv('/content/Problem3.csv')

categorical_cols = lrl_df.select_dtypes(include=['object', 'bool']).columns.tolist() # Select all obj and bool columns

for i in categorical_cols:
  n_categories = lrl_df[i].nunique() # Count the number of unique categories in each column
  print(f'Number of categories in {i}: {n_categories}')

ordinal_encoder = OrdinalEncoder()
encoded_categorical_cols = ordinal_encoder.fit_transform(lrl_df[categorical_cols]) # Turn categorical into numerical values

encoded_categorical_df = pd.DataFrame(encoded_categorical_cols, columns = categorical_cols) # Make them into a pd dataframe

numerical_df = lrl_df.drop(categorical_cols, axis =1)
encoded_df = pd.concat([numerical_df, encoded_categorical_df], axis = 1) # Concatenate into 1
encoded_df
```

```python
x = encoded_df.drop(columns=['area']) # Get all features
y = encoded_df['area'] # Get target value

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 7)

model_xg_lrl = xgb.XGBRegressor(seed = 7, learning_rate = 0.01, n_estimators = 102, max_depth = 3)
model_xg_lrl.fit(x_train, y_train)

y_pred = model_xg_lrl.predict(x_test)

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'Mean Absolute Error: {mae}')
```

# **XGBoost For Classification**

---



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import xgboost as xgb
from sklearn.metrics import mean_squared_error, mean_absolute_error, accuracy_score
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
```

```python
class_df = pd.read_csv('/content/Problem4.csv')
class_df

x = class_df.drop(columns=['Target']) # Get all features
y = class_df['Target'] # Get target value

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 7)

model_xg_class = xgb.XGBClassifier(seed = 7, learning_rate = 0.01, n_estimators = 102, max_depth = 3)
model_xg_class.fit(x_train, y_train)

preds = model_xg_class.predict(x_test)
test_accuracy = accuracy_score(y_test, preds)
train_accuracy = accuracy_score(y_train, model_xg_class.predict(x_train))

print(f'Test Accuracy: {test_accuracy}')
print(f'Train Accuracy: {train_accuracy}')
```
